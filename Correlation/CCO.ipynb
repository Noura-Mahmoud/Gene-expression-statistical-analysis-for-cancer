```python
import pandas as pd
healthy = pd.read_csv('Desktop/Statistics/lusc-rsem-fpkm-tcga_paired.txt', sep='\t')
cancer = pd.read_csv('Desktop/Statistics/lusc-rsem-fpkm-tcga-t_paired.txt', sep='\t')
```

```python
pd.options.display.max_columns = None
```

```python
healthy
```

```python
cancer
```

```python
fhealthy = healthy[healthy.astype(bool).sum(1)>27]
```

```python
fcancer = cancer[healthy.astype(bool).sum(1)>27]
```

```python
fhealthy
```

```python
fhealthy = fhealthy[cancer.astype(bool).sum(1)>27]
```

```python
fcancer = fcancer[cancer.astype(bool).sum(1)>27]
```

```python
fhealthy
```

```python
CCo = []
```

```python
from scipy.stats import pearsonr
for i in range (17275):
    G_h = fhealthy.iloc[i, 2:]
    G_c = fcancer.iloc[i, 2:]
    r, _ = pearsonr(G_h, G_c)
    CCo.append(r)
```

```python
CCo
```

```python
Max_CCo = max(CCo) 
Max_CCo
```

```python
Min_CCo = min(CCo)
Min_CCo
```

```python
max_index = CCo.index(Max_CCo)
max_index
```

```python
min_index = CCo.index(Min_CCo)
min_index
```

```python
fhealthy.iloc[ max_index , : ]
```

```python
fhealthy.iloc[ min_index , : ]
```

```python
fhealthy.insert(2, "CCo", CCo , True) 
```

```python
fcancer.insert(2, "CCo", CCo , True) 
```

```python
fhealthy
```

```python
fcancer
```

```python
fhealthy.sort_values('CCo')
```

```python
import matplotlib.pyplot as plt
plt.title('Max Correlation Coefficient Plot') 
plt.xlabel('Healthy')
plt.ylabel('Cancer')
plt.scatter(fhealthy.iloc[max_index, 3:] , fcancer.iloc[max_index, 3:])
plt.show()
```

```python
plt.title('Min Correlation Coefficient Plot') 
plt.xlabel('Healthy')
plt.ylabel('Cancer')
plt.scatter(fhealthy.iloc[min_index, 3:] , fcancer.iloc[min_index, 3:])
plt.show()
```
